#### Plotting the Data

```python
import matplotlib.pyplot as plt 
```

```python
df_comp.spx.plot(figsize=(20,7),title="S&P500 Prices")
plt.show()
```

```python
df_comp.ftse.plot(figsize=(20,7),title="FTSE100 Prices")
plt.show()
```

```python
df_comp.spx.plot(figsize=(20,7),title="S&P500 Prices")
df_comp.ftse.plot(figsize=(20,7),title="FTSE100 Prices")
plt.title("S&P VS FTSE")
plt.legend()
plt.show()
```
